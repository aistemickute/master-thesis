```python
from PIL import Image
import sys
import glob
from PIL import ImageOps
import numpy as np
import os


subdirs = os.listdir("data/Brussels statistical sectors")

for subdir in subdirs:
    filePaths = glob.glob("data/Brussels statistical sectors/" + subdir + "/*.png") #search for all png images in the folder
    for filePath in filePaths: 
        image=Image.open(filePath)
        width, height = image.size
        original_size = image.size
        image = image.crop((0, 0, width, 700)) 
        image.load()
        imageSize = image.size
    
        # remove alpha channel
        invert_im = image.convert("RGB") 
    
        # invert image (so that white is 0)
        imageBox = invert_im.getbbox()
    
        cropped=invert_im.crop(imageBox)
        if ((np.array(cropped) != 0).sum() > 90000):
            old_size = cropped.size
            new_size = (300, 300)
            new_im = Image.new("RGB", new_size, color = (0, 0, 0) )  
            new_im.paste(cropped, ((new_size[0]-old_size[0])//2,
                                  (new_size[1]-old_size[1])//2))

            imageSize = new_im.size
            print (filePath, "Size:", original_size, "New Size:", imageSize)
            new_im.save("data/Brussels statistical sectors cropped 2/" + subdir + "/" + filePath.split("\\",1)[1] )
```
